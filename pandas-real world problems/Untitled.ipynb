```python
combined_df.tail()
cdf = (combined_df.sort_values(by=["Order ID"]).reset_index(drop=True)).copy()
cdf = cdf.drop(cdf[cdf["Order ID"]=="Order ID"].index)
```

```python
cdf.tail()
```

```python
month_list =  [ "January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December" ]
cdf["Month"]= cdf["Order Date"].str[:2].astype(int)
cdf["Month"] = cdf["Month"].apply(lambda x: month_list[x-1])
```

```python
cdf.sample()
```

```python
type(cdf["Price Each"])
cdf["Price Each"].tail()
```

```python
cdf["Price Each"] = cdf["Price Each"].astype(float)
cdf["Quantity Ordered"] = cdf["Quantity Ordered"].astype(int)
```

```python
cdf.head()
```

```python
cdf["Price Each"].sum()
```

```python
cdf["Quantity Ordered"].sum()
```

```python
cdf["Expense"] = cdf["Price Each"]*cdf["Quantity Ordered"] 
```

```python
cdf.head()
```

```python
x,y = [],[]
for i in cdf["Month"].unique():
    df = cdf[cdf["Month"] == i]
    x.append(i)
    y.append(df["Expense"].sum())
print(type(x),type(y))    
```

```python
maxi = max(y)
mini = min(y)

print(maxi,x[y.index(maxi)],"\n",mini,x[y.index(mini)])
```

```python
mlt.bar(x,y)
mlt.xticks(rotation ="90")
mlt.title("Months and their Profits")
```

```python
def address(x):
    y = x.split(",")
    z = y[2].strip().split(" ")
    return (f'{y[1]} ({z[0]})')
cdf["City"] = cdf["Purchase Address"].apply(lambda x : address(x))
```

```python
cdf.head()
```

```python
cdf["City"].unique()
```

```python
city = []
for i in cdf["City"].unique():
    df = cdf[cdf["City"] == i]
    no = df["Quantity Ordered"].sum()
    consume = df["Expense"].sum()
    city.append([i,no,consume])
```

```python
city = pd.DataFrame(city, columns=["City","No of Products","Total Profit"])
city.sort_values(by="No of Products", ascending=False).reset_index(drop=True)
```

```python
city.sort_values(by="Total Profit", ascending=False).reset_index(drop=True)
```

```python
mlt.bar(city["City"],city["Total Profit"])
mlt.xticks(rotation="90")
mlt.title("City and its consumption")
```

```python
mlt.bar(city["City"],city["No of Products"])
mlt.xticks(rotation="90")
mlt.title("City and number of products sold")
```

```python
df = cdf[cdf["Order ID"].duplicated(keep=False)]
#df["Products"] = df.groupby("Order ID")["Product"].transform(lambda x:",".join(x))
df["Total Products"] = df.groupby("Order ID")["Product"].transform(lambda x:",".join(x))
df = df[["Order ID","Total Products"]].drop_duplicates()
```

```python
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for i in df["Total Products"]:
    row_list = i.split(",")
    count.update(Counter(combinations(row_list,2)))

#print(count) 
count.most_common(10)
```
