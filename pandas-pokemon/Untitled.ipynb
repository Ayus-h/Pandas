```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Pokemon.csv")
def megacheck(x):
    if "MEGA " in x.upper():
        return True
    return False
df["Mega?"]=df["Name"].apply(lambda x:megacheck(x))
```

```python
df.head()
```

```python

nonmega = df[df["Mega?"]==False][df["Legendary"] == False].sort_values("Total",ascending =0)
plt.xticks(rotation = "90")
plt.hist(nonmega["Type 1"][:20])


```
